```r
df <- read.csv("http://cardsorting.net/tutorials/25.csv")
df <- subset(df, select = -c(Uniqid, Startdate, Starttime, Endtime, QID, Comment))
```

```r
l <- subset(df, select=-(Category))
```

```r
l <- sapply(l, as.numeric)
hist(l)
```

```r
dst <- dist(t(l))
dst
```

```r
dst <- as.matrix(dst)
heatmap(dst, scale="column")
```

```r
library(qgraph)
qgraph(dst, layout='spring', vsize=3)
```
